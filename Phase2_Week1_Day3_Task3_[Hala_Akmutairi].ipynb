````python
import pymssql
import pandas as pd
conn = pymssql.connect(server='SQL6020.site4now.net', database='db_a34f35_coded' ,user='db_a34f35_coded_admin', password='coded123*')
SQL_statement="SELECT * FROM Customers"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

###2 Retrieve the names and prices of all products.

````python
SQL_statement="SELECT ProductName , Price  FROM products"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

###3 Retrieve the name and unit of products priced at 10 or less.

````python
SQL_statement="SELECT ProductName , Unit  FROM products where Price<= 10"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

###4 Select all products with a price ranging from 5 to 15.

````python
SQL_statement="SELECT *  FROM products where Price BETWEEN 5 and 15 "
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

### 5 Calculate the average price of all products.

````python
SQL_statement="SELECT AVG(price) FROM products"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

###6 Count the total number of products where SupplierID is 3.

````python
SQL_statement="SELECT COUNT(*) FROM products WHERE SupplierID=3"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

###7 Retrieve the name and price of all products priced at $20 or more, sorted first by name (in descending order) and then by price (in ascending order).

````python
SQL_statement="SELECT ProductName , Price FROM products WHERE Price>=20 ORDER BY ProductName DESC, Price ASC"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

### 8 Select all Customers records where the City is not Berlin and the Country is not Finland.

````python
SQL_statement="SELECT  * FROM Customers WHERE City !='Berlin' AND Country!='Finland'"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

### 9 Select all Customers records where the Country is Germany and the PostalCode is 12209.

````python
SQL_statement="SELECT  * FROM Customers WHERE PostalCode =12209 AND Country ='Germany'"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

### 10 Retrieve all Customers records where the City is either Berlin, London, or Seattle.

````python
SQL_statement="SELECT  * FROM Customers WHERE City IN ('Berlin','London','Seattle')"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

### 11 List all unique values from the Country column in the Customers table.

````python
SQL_statement="SELECT  DISTINCT Country FROM Customers "
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

### 12 Select all products whose name starts with the letter 'A'.

````python
SQL_statement="SELECT * FROM products WHERE ProductName LIKE 'A%'"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````

###13 Count the number of Customers in each City.

````python
SQL_statement="SELECT COUNT(*) AS Count ,City FROM Customers GROUP BY City"
df_products = pd.read_sql_query(SQL_statement, conn)
df_products
````
